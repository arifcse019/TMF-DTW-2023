```python
from matplotlib import pyplot as plt
```

```python
# Data for the plot
labels = ['NO3N', 'NH3N', 'Total Nitrogen']
values = [df['NO3N'].mean(), df['NH3N'].mean(), average_nitrogen]

# Create the plot
plt.figure(figsize=(10,6))
plt.bar(labels, values, color=['blue', 'green', 'red'])
plt.title('Average Nitrogen Values')
plt.ylabel('Value')
plt.xlabel('Nitrogen Source')
plt.show()
```
